```python
data = pd.read_csv('compTimes.log')
dataClean = data.dropna()
data.head()
```

```python
data.dropna().groupby('optLevel').agg({'compTime':['mean','std'],'solutionLength':['mean']})
#data.dropna().groupby('optLevel').agg({'compTime':['mean','std'],'solutionLength':['mean']}).to_latex()
```

```python
dataClean.sort_values('solutionLength').loc[:,['optLevel','solutionLength','compTime']]
```

```python
stdData = data.loc[data.loc[:,'file'].str.contains(r'Bomb')]
stdDataClean = stdData.dropna()

ownData = data.drop(stdData.index,axis=0)
ownDataClean = ownData.dropna()
```

```python
lowDataClean = dataClean.where(dataClean.loc[:,'solutionLength'] < 14)
highDataClean = dataClean.where(dataClean.loc[:,'solutionLength'] >= 14)
lowStdDataClean = stdDataClean.where(stdDataClean.loc[:,'solutionLength'] < 14)
highStdDataClean = stdDataClean.where(stdDataClean.loc[:,'solutionLength'] >= 14)
lowOwnDataClean = ownDataClean.where(ownDataClean.loc[:,'solutionLength'] < 14)
highOwnDataClean = ownDataClean.where(ownDataClean.loc[:,'solutionLength'] >= 14)
```

```python
data.sort_values(['file','optLevel']).groupby('optlevel')
```

```python
#Taken from https://stackoverflow.com/questions/15465645/plotting-results-of-pandas-groupby#15466103
# plt.subplot(1, 2, 1)
# for key, grp in stdDataClean.sort_values('solutionLength').groupby(['optLevel']):
#     label = "O" + str(int(key))
#     plt.plot(grp['solutionLength'],grp['compTime'], label = label)
#     plt.xlabel('SolutionLength')
#     plt.ylabel('CPU sec(s)')
# plt.legend(loc='best')
# plt.grid()

# plt.subplot(1, 2, 2)
for key, grp in stdDataClean.sort_values('solutionLength').groupby(['optLevel']):
    label = "O" + str(int(key))
    plt.semilogy(grp['solutionLength'],grp['compTime'], label = label)
    plt.xlabel('SolutionLength')
    plt.ylabel('CPU sec(s)')
plt.legend(loc='best')

plt.grid(True, which="both")
plt.show()
```

```python
#Taken from https://stackoverflow.com/questions/15465645/plotting-results-of-pandas-groupby#15466103
for key, grp in lowStdDataClean.sort_values('solutionLength').groupby(['optLevel']):
    label = "O" + str(int(key))
    plt.plot(grp['solutionLength'],grp['compTime'], label=label)
    plt.xlabel('SolutionLength')
    plt.ylabel('CPU sec(s)')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
#Taken from https://stackoverflow.com/questions/15465645/plotting-results-of-pandas-groupby#15466103
for key, grp in highStdDataClean.sort_values('solutionLength').groupby(['optLevel']):
    label = "O" + str(int(key))
    plt.plot(grp['solutionLength'],grp['compTime'], label=label)
    plt.xlabel('SolutionLength')
    plt.ylabel('CPU sec(s)')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
#Taken from https://stackoverflow.com/questions/15465645/plotting-results-of-pandas-groupby#15466103
fig = plt.figure()
for key, grp in lowOwnDataClean.sort_values('file').groupby(['optLevel']):
    label = "O" + str(key)
    plt.plot(grp['file'],grp['compTime'], label=label)
    plt.xlabel('File')
    plt.ylabel('CPU sec(s)')
plt.legend(loc='best')

for tick in fig.get_xticklabels():
    tick.set_rotation(45)
plt.grid()
plt.show()
```

```python
#Taken from https://stackoverflow.com/questions/15465645/plotting-results-of-pandas-groupby#15466103
for key, grp in highOwnDataClean.sort_values('solutionLength').groupby(['optLevel']):
    plt.bar(grp['solutionLength'],grp['compTime'], label=key)
    plt.xlabel('SolutionLength')
    plt.ylabel('CPU sec(s)')
plt.legend(loc='best')
plt.grid()
plt.show()
```
